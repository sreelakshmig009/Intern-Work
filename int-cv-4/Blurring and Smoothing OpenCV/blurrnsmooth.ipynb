```python
#import the requirements
import cv2
from google.colab.patches import cv2_imshow
```

```python
#read the image 
image = cv2.imread('Images/devincept.jpg')
cv2_imshow(image)  
```

#**Averaging**

```python
# kernel size
ksize = (8, 8)

# Using cv2.blur() method for averaging
avg = cv2.blur(image, ksize) 
cv2_imshow(avg)
```

```python
#increasing kernel size increases the blurriness of the image
ksize = (15,15)

avg = cv2.blur(image, ksize) 
cv2_imshow(avg)
```

#**Gaussian Blurring**

```python
#kernel size
ksize = (9,9)

# Using cv2.GaussianBlur() for blurring the image
blur = cv2.GaussianBlur(image,ksize,0)
cv2_imshow(blur)
```

#**Median Blurring**

```python
median = cv2.medianBlur(image,5)
cv2_imshow(image)
```

#**Bilateral Filtering**

```python
blur1 = cv2.bilateralFilter(image,9,75,75)
cv2_imshow(blur1)
```
